```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#converting 0 and 1 values of senior citizen to yes/no to make it easier to understand
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen']=df["SeniorCitizen"].apply(conv)
        
```

```python
df.head(7)
```

```python
#Count Customers by churn
ax=sns.countplot(x="Churn", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

IN CONCLUSION 26.54% of Customers have churned out

```python
plt.figure(figsize=(3,3))
sns.countplot(x="gender",hue="Churn",data=df)
plt.title("Churn by gender")
plt.show()
```

Churn in both the genders is almost the same

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x="SeniorCitizen",data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
#Calculate percentages
total_counts = (
    df.groupby('SeniorCitizen')['Churn']
      .value_counts(normalize=True)
      .unstack() * 100
)

# Create plot
ax = total_counts.plot(kind='bar', stacked=True, figsize=(4,4))

# Add percentage labels
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:  # Avoid printing labels for zero-height bars
        ax.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_y() + height/2,
            f'{height:.1f}%',
            ha='center',
            va='center',
            fontsize=9
        )

plt.title("Churn by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")

plt.legend(title="Churn")
plt.xticks(rotation=0)
plt.show()
```

Comparatively greater % of people have churned out in the senior citizen category

```python
plt.figure(figsize=(9,6))
sns.histplot(x="tenure",hue="Churn",data=df,bins=72)
plt.show()
```

People who have used the telecom services long enough have stayed and mostly people who have used it from 1-2 months have churned out. This indicates the importance of improving customer onboarding and early engagement strategies for customer retention

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="Contract",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

People who have month to month contract are likely to churn out than people who have 1-2 years of contract

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The majority of customers who do not churn tend to have services like PhoneService, InternetService (especially DSL) and Online Security enabled.
For services like Online Backup, Tech Support and Streaming TV churn rates are noticeably high when not used

```python
plt.figure(figsize = (6,4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

Customers are likely to churn when they are using electronic check as payment

KPI SECTION:

```python
total_customers = df.shape[0]

churned_customers = df[df['Churn'] == 'Yes'].shape[0]

retained_customers = df[df['Churn'] == 'No'].shape[0]

churn_rate = (churned_customers / total_customers) * 100

retention_rate = (retained_customers / total_customers) * 100

avg_tenure = df['tenure'].mean()

avg_monthly_charges = df['MonthlyCharges'].mean()

avg_total_charges = df['TotalCharges'].mean()

print(f"Total Customers        : {total_customers}")
print(f"Churned Customers      : {churned_customers}")
print(f"Retained Customers     : {retained_customers}")
print(f"Churn Rate             : {churn_rate:.2f}%")
print(f"Retention Rate         : {retention_rate:.2f}%")
print(f"Average Tenure         : {avg_tenure:.2f} months")
print(f"Average Monthly Charge : {avg_monthly_charges:.2f}")
print(f"Average Total Charges  : {avg_total_charges:.2f}")
```
